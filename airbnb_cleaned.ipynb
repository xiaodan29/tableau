```python
import pandas as pd
import numpy as np
```

```python
#load Airbnb xlsx
airbnb_df = pd.read_excel("airbnb.xlsx")
airbnb_df.head(20)
```

```python
airbnb_df.columns
```

```python
airbnb_df.dtypes
```

```python
# Check for null values in each column
print(airbnb_df.isnull().sum())
```

```python
# Delete the column "Zipcode" with Null for further use in Tableau
Airbnb=airbnb_df.dropna(subset=['Zipcode'])
print(Airbnb.isnull().sum())
```

```python
# Delete the column "Beds" with no detailed info 
airbnb=Airbnb.dropna(subset=['Beds'])
print(airbnb.isnull().sum())
```

```python
# Convert column 'Zipcode' from float64 to int
airbnb['Zipcode'] = airbnb['Zipcode'].astype(int)
```

```python
airbnb['Zipcode']
```

```python
# Delete rows with Zipcode with the zipcode are not in New Yord City
airbnb = airbnb[~airbnb['Zipcode'].isin([99135, 94103, 0o1003, 0o7712, 0o7105, 0o7030])]
```

```python
# Export this cleaned aibnb data to a CSV file
airbnb.to_csv('airbnb_cleaned.csv', index=False)
```
